### Chap 1.4.4 Smoothing Images

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

##### 2D Convolution (Image Filtering)

```python
img=cv2.imread('opencv.png')

kernel=np.ones((5,5),np.float32)/25
dst=cv2.filter2D(img,-1,kernel)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
img=cv2.imread('game.png')

kernel=np.ones((5,5),np.float32)/25
dst=cv2.filter2D(img,-1,kernel)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]),plt.yticks([])
plt.show()
```

##### Image Blurring (Image Smoothing)

1. Averaging

```python
img=cv2.imread('opencv.png')

blur = cv2.blur(img,(5,5))

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img=cv2.imread('game.png')

blur3 = cv2.blur(img,(3,3))
blur5 = cv2.blur(img,(5,5))

plt.subplot(121),plt.imshow(blur3),plt.title('blur3')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur5),plt.title('Blur5')
plt.xticks([]), plt.yticks([])
plt.show()
```

2. Gaussian Filtering

```python
img=cv2.imread('game.png')

blur = cv2.blur(img,(15,15))
Gaussianblur = cv2.GaussianBlur(img,(5,5),0)

plt.subplot(121),plt.imshow(blur),plt.title('blur')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(Gaussianblur),plt.title('GaussianBlur')
plt.xticks([]), plt.yticks([])
plt.show()
```

3. Median Filtering

```python
img=cv2.imread('game.png')

blur = cv2.blur(img,(15,15))
median = cv2.medianBlur(img,5)

plt.subplot(121),plt.imshow(blur),plt.title('blur')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('median')
plt.xticks([]), plt.yticks([])
plt.show()
```

4. Bilateral Filtering

```python
img=cv2.imread('game.png')

blur = cv2.blur(img,(15,15))
bilateralFilter = cv2.bilateralFilter(img,9,75,75)

plt.subplot(121),plt.imshow(blur),plt.title('blur')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(bilateralFilter),plt.title('bilateralFilter')
plt.xticks([]), plt.yticks([])
plt.show()
```
