## Chap 1.5.10 Feature Matching + Homography to find Objects

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

##### Feature Matching

```python
img1 = cv2.imread('view1.png',0)   # queryImage
img2 = cv2.imread('view5.png',0) # trainImage

sift = cv2.SIFT_create()
kp1, des1 = sift.detectAndCompute(img1,None)
kp2, des2 = sift.detectAndCompute(img2,None)

# create BFMatcher object
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

# Match descriptors.
#matches = bf.match(des1,des2)
matcher = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
matches = matcher.match(des1, des2)

# Sort them in the order of their distance.
matches = sorted(matches, key = lambda x:x.distance)

# Draw first 10 matches.
#img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:10], flags=2)
img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches, None, flags=cv2.DRAW_MATCHES_FLAGS_NOT_DRAW_SINGLE_POINTS)
plt.imshow(img3),plt.show()
```
